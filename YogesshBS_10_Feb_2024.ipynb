## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
ECom_Data['Revenue'] = 0
for i in range(len(ECom_Data)):
    if ECom_Data.loc[i, 'discounted_price'] > 600:
        ECom_Data.loc[i, 'Revenue'] = ECom_Data.loc[i, 'discounted_price'] * 0.25
    elif ECom_Data.loc[i, 'discounted_price'] > 350 and ECom_Data.loc[i, 'discounted_price'] <= 600:
        ECom_Data.loc[i, 'Revenue'] = ECom_Data.loc[i, 'discounted_price'] * 0.15
    elif ECom_Data.loc[i, 'discounted_price'] > 100 and ECom_Data.loc[i, 'discounted_price'] <= 350:
        ECom_Data.loc[i, 'Revenue'] = ECom_Data.loc[i, 'discounted_price'] * 0.10
    else:
        ECom_Data.loc[i, 'Revenue'] = ECom_Data.loc[i, 'discounted_price']*0.05

total_revenue = ECom_Data['Revenue'].sum()
print(f"Total Revenue: {total_revenue}")
#print(ECom_Data)
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


### Ans 6

```python
ECom_Data['Brand_Revenue'] = 0
for i in range(len(ECom_Data)):
    ECom_Data.loc[i,'Brand_Revenue'] = ECom_Data.loc[i,'discounted_price']-ECom_Data.loc[i,'Revenue']

Brand_Revenue=ECom_Data['Brand_Revenue'].sum()
print(f'Total Brand Revenue: {Brand_Revenue}')
print("\nTop 10 Brands having maximum Revenue")
ECom_Data.groupby('Brand')['Brand_Revenue'].sum().sort_values(ascending=False).head(10)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
# Group the data by region and month
ECom_Data['Order_Date']= pd.to_datetime(ECom_Data['Order_Date'],dayfirst=True)
grouped_data = ECom_Data.groupby(['Region', pd.to_datetime(ECom_Data['Order_Date']).dt.strftime('%Y-%m')])['Revenue'].sum().reset_index()
#print(grouped_data)
# Create a lineplot for each region
for region in grouped_data['Region'].unique():
    region_data = grouped_data[grouped_data['Region'] == region]
    plt.figure(figsize=(17,7))
    sns.lineplot(x='Order_Date', y='Revenue', data=region_data)
    plt.xlabel('Month and Year')
    plt.ylabel('Revenue')
    plt.title(f'Monthly Revenue for {region}')
    plt.xticks(rotation=45)
    plt.show()
```

### Ans 9 b)

From the above Linplot it's clear,
For East Region,
The best performing month is Decemeber,2020 and the worst performing month is October,2021
For West Region,
The best performing month is Decemeber,2020 and the worst performing month is October,2021
For South Region,
The best performing month is Decemeber,2020 and the worst performing month is May,2021
For North Region,
The best performing month is Decemeber,2020 and the worst performing month is July,2021
